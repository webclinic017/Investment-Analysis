```python
import pandas as pd
```

```python
rounds2 = pd.read_csv("data/rounds2.csv",encoding = "ISO-8859-1")
companies = df = pd.read_csv('data/companies.txt', sep = "\t",encoding = "ISO-8859-1")
```

```python
# Unique companies present in rounds2
rounds2['company_permalink'].unique().shape[0]
```

```python
# Unique companies in comapanies .txt file
companies['permalink'].unique().shape[0]
```

```python
# checking if there are any companies from round2 not present in companies
rounds2.company_permalink.isin(companies.permalink).astype(int).sum()
```

```python
# merging rounds2 to companies
master_frame = rounds2.merge(companies, how='outer', right_on='permalink',left_on='company_permalink')
```
